# 01 -- Exploratory Data Analysis

Load the cached parquet outputs and explore the distribution of linkage density and capability components across communities and regions.

```python
import polars as pl
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid')
df = pl.read_parquet('../results/capability_scores.parquet')
df.head()
```

```python
df.describe()
```

```python
pdf = df.to_pandas()
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(pdf['linkage_density'], bins=40, ax=axes[0])
axes[0].set_title('Linkage density distribution')
sns.histplot(pdf['capability_score'], bins=40, ax=axes[1])
axes[1].set_title('Capability score distribution')
plt.tight_layout()
```

```python
if 'region' in pdf.columns:
    sns.boxplot(data=pdf, x='region', y='capability_score')
    plt.title('Capability by region')
```
