# 03 -- Causal DAG

We posit the following DAG:

```
   Funding/Horizon ---
                      |
                      v
  Community size --> Linkage density --> Capability score
                      ^                   ^
         Region ------+-------------------+
```

**Identification strategy:** condition on community size, region, discipline diversity, OA share. Use EU Horizon exposure as an instrument for linkage density (exogeneity: Horizon mandates predate most metadata linking; exclusion: mandates affect capability only through the infrastructural participation they induce).

````python
import networkx as nx, matplotlib.pyplot as plt
G = nx.DiGraph()
G.add_edges_from([
    ('Funding/Horizon', 'Linkage density'),
    ('Community size', 'Linkage density'),
    ('Community size', 'Capability'),
    ('Region', 'Linkage density'),
    ('Region', 'Capability'),
    ('Linkage density', 'Capability'),
])
pos = nx.spring_layout(G, seed=1)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2500, arrowsize=18)
plt.title('Assumed causal DAG')
plt.show()
````
